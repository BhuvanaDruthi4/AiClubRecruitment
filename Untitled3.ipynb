```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings


from sklearn.metrics import mean_absolute_percentage_error
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor
```

```python
demand_df = pd.read_excel("PGCB_date_power_demand.xlsx")
demand_df["datetime"] = pd.to_datetime(demand_df["datetime"],errors="coerce")
demand_df = demand_df.set_index("datetime")
demand_df = demand_df[~demand_df.index.isna()]
demand_df = demand_df[~demand_df.index.duplicated()]
demand_df = demand_df.sort_index()

demand_df = demand_df.resample("h").mean(numeric_only=True)

demand_df = demand_df.ffill()
demand_df = demand_df.bfill()


time_gaps = demand_df.index.to_series().diff().dropna()


Q1 = demand_df['demand_mw'].quantile(0.25)
Q3 = demand_df['demand_mw'].quantile(0.75)
IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR



outlier_mask = (demand_df['demand_mw'] < lower_bound) | (demand_df['demand_mw'] > upper_bound)


median_value = demand_df['demand_mw'].median()
demand_df.loc[outlier_mask, 'demand_mw'] = median_value


demand_df['hour']        = demand_df.index.hour
demand_df['day_of_week'] = demand_df.index.dayofweek
demand_df['month']       = demand_df.index.month
demand_df['is_weekend']  = (demand_df.index.dayofweek >= 5).astype(int)

demand_df['lag_1']   = demand_df['demand_mw'].shift(1)
demand_df['lag_24']  = demand_df['demand_mw'].shift(24)
demand_df['lag_168'] = demand_df['demand_mw'].shift(168)
demand_df['target'] = demand_df['demand_mw'].shift(-1)
demand_df = demand_df.dropna(
    subset=["lag_1", "lag_24", "lag_168","target"]
)

```

```python
weather_df = pd.read_excel(
    "weather_data.xlsx",
    header=3
)

weather_df["time"] = pd.to_datetime( weather_df["time"], errors="coerce")
weather_df = weather_df.set_index('time')
weather_df = weather_df.sort_index()
weather_df = weather_df[~weather_df.index.duplicated()]
weather_df = weather_df.ffill()
weather_df = weather_df.bfill()

```

```python

econ_df = pd.read_csv("economic_full_1.csv")
needed_indicators = ["GDP (current US$)","GDP growth (annual %)","Population, total","Inflation, consumer prices (annual %)",
"Electric power consumption (kWh per capita)"]

econ_df = econ_df[ econ_df["Indicator Name"].isin(needed_indicators)]
econ_df = econ_df.melt( id_vars=["Country Name","Indicator Name","Indicator Code"],
    var_name="year",
    value_name="value"
)



econ_df = econ_df.pivot(index="year",columns="Indicator Name",values="value")
econ_df.index = pd.to_datetime(econ_df.index)


```

```python
data = demand_df.join(weather_df, how="left")
data['year'] = data.index.year

econ_df.index = econ_df.index.year.astype(int)
econ_df.index.name = 'year'
econ_df = econ_df.reset_index()

data = data.reset_index().merge(econ_df, on='year', how='left')
data = data.set_index('datetime')
data = data.drop(columns=['year'])

data = data.ffill()
data = data.bfill()



train_data = data[data.index.year <= 2023]
test_data = data[data.index.year == 2024]
y_train = train_data['target']
y_test  = test_data['target']


X_train = train_data.drop(columns=['demand_mw', 'target'])
X_test  = test_data.drop(columns=['demand_mw', 'target'])




X_train = X_train.select_dtypes(include=["number"])

X_test = X_test.select_dtypes(include=["number"])



X_train = X_train.fillna(0)
X_test = X_test.fillna(0)



lr = LinearRegression()

lr.fit(X_train, y_train)

lr_pred = lr.predict(X_test)

rf = RandomForestRegressor(
    n_estimators=40,
    max_depth=10,
    random_state=42,
    n_jobs=-1
)

rf.fit(X_train, y_train)

rf_pred = rf.predict(X_test)


mape_lr = mean_absolute_percentage_error(y_test, lr_pred) * 100
mape_rf = mean_absolute_percentage_error(y_test, rf_pred) * 100

print(f"Linear Regression MAPE: {mape_lr:.2f}%")
print(f"Random Forest     MAPE: {mape_rf:.2f}%")



feat_imp = pd.Series(rf.feature_importances_, index=X_train.columns)

plt.figure(figsize=(10, 6))
feat_imp.nlargest(15).sort_values().plot(kind='barh', color='blue')
plt.title('Top 15 Feature Importances - Random Forest')
plt.xlabel('Importance Score')
plt.tight_layout()
plt.show()

print("Top 10 most important features:")
print(feat_imp.nlargest(10))

plt.figure(figsize=(14, 4))
plt.plot(y_test.values[:500], label='Actual', color='blue')
plt.plot(rf_pred[:500], label='Predicted (RF)', color='red', alpha=0.7)
plt.title('Actual vs Predicted Demand - Random Forest')
plt.legend()
plt.tight_layout()
plt.show()

```
